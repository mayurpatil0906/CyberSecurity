```python
import psutil
import time
import logging

# Configure logging to write to both console and a log file
logging.basicConfig(
    format='%(asctime)s - %(message)s',
    level=logging.INFO,
    handlers=[logging.StreamHandler(), logging.FileHandler("cyberSecurity.log", mode='a')]
)

# Monitor CPU, Memory, Disk Usage
def monitor_system():
    cpu_usage = psutil.cpu_percent(interval=1)
    memory_usage = psutil.virtual_memory().percent
    disk_usage = psutil.disk_usage('/').percent
    net_stats = psutil.net_io_counters()

    logging.info("========== SYSTEM STATS ==========")
    logging.info(f"CPU Usage       : {cpu_usage}%")
    logging.info(f"Memory Usage    : {memory_usage}%")
    logging.info(f"Disk Usage      : {disk_usage}%")
    logging.info(f"Bytes Sent      : {net_stats.bytes_sent}")
    logging.info(f"Bytes Received  : {net_stats.bytes_recv}")
    logging.info("===================================")
    
    # Flush logs immediately to the file
    for handler in logging.getLogger().handlers:
        if isinstance(handler, logging.FileHandler):
            handler.flush()

# Monitor for a fixed number of cycles (e.g., 5 cycles) or until some condition
if __name__ == "__main__":
    while True:
        monitor_system()
        time.sleep(2)
        
# if __name__ == "__main__":
#     cycle_count = 5  # Number of cycles to run before stopping
#     for i in range(cycle_count):
#         monitor_system()
#         time.sleep(2)  # Wait for 2 seconds between each monitoring cycle

#     logging.info("Monitoring complete. System data saved to 'cybersecurity.log'.")
```
